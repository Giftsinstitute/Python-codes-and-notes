```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gdp = pd.read_csv('RealGDP17092022.csv')
gdp.head()
```

```python
gdp.reset_index(inplace=True)
```

```python
gdp.columns
```

```python
gdp.rename({'index':'Year', 'tyear':'Period', 'Period':'Agriculture', 'Agriculture':'CropProduction', 'CropProduction':'Livestock',
       'Livestock':'Forestry', 'Forestry':'Fishing', 'Fishing':'Industry', 'Industry':'MiningAndQuarrying', 'MiningAndQuarrying':'CrudePetroleumNaturalGas',
       'CrudePetroleumAndNaturalGas':'CoalMining', 'CoalMining':'MetalOres', 'MetalOres':'QuarryingAndOtherMinerals',
       'QuarryingAndOtherMinerals':'Manufacturing', 'Manufacturing':'OilRefining', 'OilRefining':'Cement', 'Cement':'FoodBeverageAndTobacco',
       'FoodBeverageAndTobacco':'TextileApparelAndFootwear', 'TextileApparelAndFootwear':'WoodAndWoodProducts',
       'WoodAndWoodProducts':'PulpPaperAndPaperProducts', 'PulpPaperAndPaperProducts':'ChemicalandPharmaceuticalProducts',
       'ChemicalAndPharmaceuticalProducts':'NonMetallicProducts', 'NonMetallicProducts':'PlasticandRubberProducts',
       'PlasticAndRubberProducts':'ElectricAndElectronics', 'ElectricalAndElectronics':'BasicMetalIronAndSteel',
       'BasicMetalIronAndSteel':'VehicleAss', 'MotorVehiclesAndAssembly':'OtherManufacturing',
       'OtherManufacturing':'ElectricityGassteam', 'ElectricityGasSteamAndAirCon':'WaterSupply',
       'WaterSupplySewageWaste':'Construction', 'Construction':'Services', 'Services':'Trade', 'Trade':'Accommodation',
       'AccommodationAndFoodServices':'TRansportation', 'TransportationAndStorage':'RoadTransport',
       'RoadTransport':'RailTransport', 'RailTransportAndPipelines':'WaterTransport', 'WaterTransport':'AirTransport',
       'AirTransport':'Transpoerservices', 'TransportServices':'PostAndCourier', 'PostAndCourierServices':'InformationAndComm',
       'InformationAndCommunication':'Publishing', 'Publishing':'MotionPictures',
       'MotionPicturesSoundRecordingAndMusicProduction':'Broadcasting', 'Broadcasting':'ArtsEntertainment',
       'ArtsEntertainmentAndRecreation':'FinanceAndInsurance', 'FinanceAndInsurance':'FinancialInstitution',
       'FinancialInstitutions':'Insurance', 'Insurance':'RealEstate', 'RealEstate':'SciTech',
       'ProfessionalScientificAndTechnicalServices':'AdminSupport',
       'AdministrativeAndSupportServicesBusinessServices':'PunblicAdmin',
       'PublicAdministration':'Education', 'Education':'HumanHEalth', 'HumanHealthAndSocialServices':'Otherservices',
       'OtherServices':'GDPatBasicprice', 'GDP at 2010 Constant Basic Prices':'NetTaxes',
       'NetTaxesOnProducts':'GDPatMarket Prices', 'GDP at 2010 Constant Market Prices':'NILL'},axis = 1,inplace =True)
gdp.head()
```

```python
gdp2 = gdp[gdp['Period'] == 'Annual']
```

```python
gdp2
```

```python
plt.style.use('ggplot')
plt.plot(gdp2['Year'],gdp2['Agriculture'],ls = '--',lw=2.5,marker='o',mfc='purple',markevery =5,color ='green')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('VISUALIZING THE GROWTH OF THE GDP OF AGRICULTURE INDUSTRY IN MILLIONS')
plt.show()
```

```python
fig = plt.figure(figsize=(15,15))
axes = fig.add_axes([0.1,0.1,0.6,0.6])#left,bottom,width,height
axes.plot(gdp2['Year'],gdp2['Agriculture'],ls = '--',lw=2.5,marker='o',mfc='purple',markevery =5,color ='green')
axes.set_xlabel('Year',color = 'green',fontsize = 25)
axes.set_ylabel('GDP',color = 'green',fontsize = 25)
axes.set_title('VISUALIZING THE GROWTH OF THE GDP OF AGRICULTURE INDUSTRY IN MILLIONS',color = 'blue',fontsize = 25)
plt.show()
```
